```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline

# 
import csv

x = []
y = []

with open('example.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))

plt.plot(x,y, label='Loaded from file!')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()

# 
x = np.random.randn(1000)
plt.hist(x, 50)
plt.show()

#
x = np.linspace(0, 2 * np.pi, 50) # 50  50  [0, 2pi] 
plt.plot(x, np.sin(x)) #  x x 
plt.plot(x, np.sin(x),
        x, np.sin(2 * x)) #
plt.show() # 

#
x = np.linspace(0, 2 * np.pi, 50)
plt.plot(x, np.sin(x), 'r-o',
        x, np.cos(x), 'g--') #'r'  'g' '-o' '--' 
plt.show()

#
x = np.linspace(0, 2 * np.pi, 50)
plt.subplot(2, 1, 1) # 
plt.plot(x, np.sin(x), 'r')
plt.subplot(2, 1, 2) # plot()  subplot() 
plt.plot(x, np.cos(x), 'g')
plt.show()

#
x = np.linspace(0, 2 * np.pi, 50)
y = np.sin(x)
plt.scatter(x,y) # plot  'bo' 
plt.show()

#
x = np.random.rand(1000)
y = np.random.rand(1000)
size = np.random.rand(1000) * 50
colour = np.random.rand(1000)
plt.scatter(x, y, size, colour)
plt.colorbar() #
plt.show()

# 
x = np.linspace(0, 2 * np.pi, 50)
plt.plot(x, np.sin(x), 'r-x', label='Sin(x)')
plt.plot(x, np.cos(x), 'g-^', label='Cos(x)')
plt.legend() # label
plt.xlabel('Rads') #  x 
plt.ylabel('Amplitude') #  y 
plt.title('Sin and Cos Waves') # 
plt.show()

#
slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
cols = ['c','m','r','b']

plt.pie(slices,
        labels=activities,
        colors=cols,
        startangle=90,
        shadow= True,
        explode=(0,0.1,0,0),
        autopct='%1.1f%%')

plt.title('Interesting Graph\nCheck it out')
plt.show()
```
